```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
recall_df = pd.read_csv('data/recall_long.csv')
precision_df = pd.read_csv('data/precision_long.csv')
f1_score_df = pd.read_csv('data/f1_long.csv')
roc_auc_df = pd.read_csv('data/rocauc_long.csv')
```

# Recall

```python
recall_df
```

```python
my_colors = ["#de1c26", "#84161a"]
sns.set_style("ticks", {"axes.facecolor": "#fcf2f2"})
f, ax = plt.subplots(1, figsize=(10, 10))
f.suptitle('Distribution and quartiles of Recall between Lifestage and Models', fontsize=15)

# Set the plot style and background color
sns.set_style("ticks", {"axes.facecolor": "#fcf2f2"})
sns.violinplot(x='Lifestage', y='recall', hue='Model',data=recall_df, split=True, inner='quartiles',
                               linewidth=1.3,
                               palette=my_colors, ax=ax)
ax.set_xlabel("Lifestage", size=14, alpha=0.8)
ax.set_ylabel("Recall", size=14, alpha=0.8)

for l in ax.lines:
    l.set_color('#fcf2f2')

# Adjust the gap between the title and the plot
plt.subplots_adjust(top=0.95)

# Get the legend object
legend = ax.legend()

# Set the background color of the legend
legend.get_frame().set_facecolor('white')
# Add title to the legend
legend.set_title('Model')

sns.despine()

plt.show()
```
