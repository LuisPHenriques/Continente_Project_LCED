```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('customer_rec2.csv')
```

```python
### Trim the white spaces
df.sort_values(by=['DIFFERENCE_PERC_TOTAL'], inplace=True)
x = df['SUBCAT_DSC_EXT']
y = df['DIFFERENCE_PERC_TOTAL']

mask = y < 0
sns.set_style("ticks", {"axes.facecolor": "#fcf2f2"})
#Plot Horizontal Bar Chart
plt.barh(df['SUBCAT_DSC_EXT'][mask], df['DIFFERENCE_PERC_TOTAL'][mask], color='#84161a')
plt.barh(df['SUBCAT_DSC_EXT'][~mask], df['DIFFERENCE_PERC_TOTAL'][~mask], color='#de1c26')
# Sort the X axis by the values in the column
#plt.xticks(list(df['SUBCAT_DSC_EXT'].astype(str)))
plt.title('Difference in % of between Recommended and Total Values')
# Rotate x-labels
plt.ylabel('Subcategory')
plt.xlabel('% Change')
## Turn the whole plot in 90 degree
#plt.xticks(rotation=90)



plt.savefig('test.png', dpi=900, bbox_inches='tight')
```
